```python
import matplotlib.pyplot as plt
import matplotlib.style
import numpy as np
```

```python
_, ax = plt.subplots()
cities = {("Seoul", "ko-"), ("Busan", "r<--"), ("Daejeon", "bd-")}
y = np.arange(2000, 2021, dtype = int)
min, max = 10 ** 9, - 10 ** 9 
for c, m in cities:
    v = np.random.randn(21).cumsum()
    ax.plot(y, v, m)
    plt.annotate(c, (2021, v[20]))
    if min > v.min():min = v.min()
    if max < v.max(): max = v.max()
plt.axis([1999,2025, min-1, max+3])
plt.title("City Growth")
plt.xlabel("year")
plt.ylabel("population")
plt.legend(cities)
plt.show()

```

```python
import random 

n = 110000

inCount = 0 
for i in range(1, n+1):
    x, y = (random.uniform(0,1), random.uniform(0,1))
    isin = (x**2 + y**2 <= 1)
    inCount += isin

print(f"pi = {inCount*4/n}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.style
import numpy as np
matplotlib.style.use("bmh")
_, ax = plt.subplots()
plt.axis((0, 1, 0, 1))
ax.set_aspect(1)
inCount = 0             # 4     
inListX, inListY = [], []
outListX, outListY = [], []
n = 5000
for i in range(1, n+1):
    x, y = (random.uniform(0, 1), random.uniform(0, 1))
    isIn = (x**2 + y**2 <= 1)
    inCount += isIn
    if isIn:
        inListX.append(x)
        inListY.append(y)
    else:
        outListX.append(x)
        outListY.append(y)
    if i%100 == 0:
        ax.set_title(f"n = {i}, pi ~ {inCount*4/i:5f}")
        ax.scatter(inListX, inListY, c='b', s=2)
        ax.scatter(outListX, outListY, c='r', s=2)
        plt.pause(0.1)
        inListX, inListY = [], []
        outListX, outListY = [], []

print(f"pi = {inCount*4/n}")
```
